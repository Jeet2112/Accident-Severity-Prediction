```python
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        
        # Return the dataframe with missing information
        return mis_val_table_ren_columns
    
    
df_missing = missing_values_table(df)
df_missing
```


# Baseline Modelling

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

oversample = SMOTE()

labels = df1.drop(['Accident_severity'], axis=1).columns[1:]

x = df1[labels]
y = df1['Accident_severity']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=35)
```

```python
x_train, y_train = oversample.fit_resample(x_train, y_train)
```

```python
'''from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))'''
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score, f1_score

xt = ExtraTreesClassifier()
xt.fit(x_train, y_train)
y_pred1 = xt.predict(x_test)
    
def performance(xt, y_test, y_pred1):
    accuracy = accuracy_score(y_test, y_pred1)
    f1 = f1_score(y_test, y_pred1, average= 'weighted')
    print("Accuracy: %.2f%%" % (accuracy * 100.0))
    print("F1 Score: %.2f%%" % (f1 * 100.0))
    
performance(xt, y_test, y_pred1)
```

```python
xt.get_params()
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold, KFold

start = time.time()
gkf = KFold(n_splits=3, shuffle=True, random_state=42).split(X=x_train, y=y_train)

params = {
    'n_estimators': range(100, 500, 100),
    'ccp_alpha': [0.0, 0.1],
    'criterion': ['gini'],
    'max_depth': [5,11],
    'min_samples_split': [2,3],
}

extree_estimator = ExtraTreesClassifier()

gsearch = GridSearchCV(
    estimator= extree_estimator,
    param_grid= params,
    scoring='f1_weighted',
    n_jobs=-1,
    cv=gkf,
    verbose=1,
)

gsearch.fit(X=x_train, y=y_train)
(gsearch.best_params_, gsearch.best_score_)
```

```python
gkf2 = KFold(n_splits=3, shuffle=True, random_state=35).split(X=x_train, y=y_train)

xt2= ExtraTreesClassifier(ccp_alpha= 0.0,
                          criterion= 'gini',
                          min_samples_split= 2,
                          n_estimators= 400)

params_2 = {
    'n_estimators': range(600, 1000, 100),
    'max_depth': [15,20],
    'min_samples_split': [2,3],
    'class_weight': ['balanced', None]
}
    
gsearch2 = GridSearchCV(
    estimator= xt2,
    param_grid= params_2,
    scoring='f1_weighted',
    n_jobs=-1,
    cv=gkf2,
    verbose=3,
)

gsearch2.fit(X=x_train, y=y_train)
(gsearch2.best_params_, gsearch2.best_score_)
```

```python
gkf3 = KFold(n_splits=3, shuffle=True, random_state=50).split(X=x_train, y=y_train)

xt3= ExtraTreesClassifier(ccp_alpha= 0.0,
                          criterion= 'gini',
                          min_samples_split= 2,
                          n_estimators= 400)

params_3 = {
    'n_estimators': range(800, 1100, 100),
    'max_depth': [18,23],
    'min_samples_split': [2,3],
    'class_weight': ['balanced', None]
}
    
gsearch3 = GridSearchCV(
    estimator= xt3,
    param_grid= params_3,
    scoring='f1_weighted',
    n_jobs=-1,
    cv=gkf3,
    verbose=3,
)

gsearch3.fit(X=x_train, y=y_train)
(gsearch3.best_params_, gsearch3.best_score_)
```

```python
xt_tuned = ExtraTreesClassifier(ccp_alpha= 0.0,
                              criterion= 'gini',
                              min_samples_split= 2,
                              n_estimators= 1000,
                              max_depth= 23,
                              class_weight= 'balanced',
                              random_state= 50)
                                
xt_tuned.fit(x_train, y_train)
y_pred_tuned = xt_tuned.predict(x_test)

f1_tuned = f1_score(y_test, y_pred_tuned, average='weighted')
print(f1_tuned)
```
